```python
# Excel-Datei einlesen
filename = "ferien60_61"
file_path = f"Daten/Schulferien/{filename}.xlsx"
df = pd.read_excel(file_path)

# Manuelle Eingabe, um Zeilen zu löschen
while True:
    display(df)
    try:
        start_index = int(input("Gib den Startindex zum Löschen ein (Negativwert zum Beenden): "))
        if start_index < 0:
            break  # Eingabe eines negativen Werts beendet die Schleife
        end_index = int(input("Gib den Endindex zum Löschen ein: "))
        
        # Zeilen im angegebenen Bereich löschen
        if start_index >= 0 and end_index < len(df):
            df = df.drop(index=range(start_index, end_index + 1))
            print(f"Zeilen von Index {start_index} bis {end_index} wurden gelöscht.")
        else:
            print("Ungültige Indexwerte. Bitte gib gültige Indizes ein.")

    except ValueError:
        print("Ungültige Eingabe. Bitte gib ganze Zahlen ein.")

# DataFrame in Excel-Datei speichern (optional)
output_file_path = f'Daten/Schulferien/EDIT/{filename}.xlsx'
df.to_excel(output_file_path, index=False)
print(f"Der aktualisierte DataFrame wurde in {output_file_path} gespeichert.")
```

# Entferne Leerzeichen am Ende der Zelle

```python
import pandas as pd
import os



# Beispieldatenframe erstellen
#df = pd.DataFrame({'Spalte1': ['Hallo Welt ', 'Hallo'], 'Spalte2': ['Test ', 'Test2']})


# Funktion zum Entfernen von Leerzeichen am Ende
def remove_trailing_spaces(df):
  """
  Durchläuft einen Dataframe und entfernt Leerzeichen am Ende von Strings in allen Spalten.
  Args:
    df: Der zu bearbeitende Dataframe.
  Rückgabe:
    Der modifizierte Dataframe.
  """
  for col in df.columns:
    # Spalte mit apply-Funktion bearbeiten
    df[col] = df[col].astype(str)
    df[col] = df[col].apply(lambda x: x[:-1] if x.endswith(' ') else x)
  return df

directory = "Daten/Schulferien"
for filename in os.listdir(directory):
  df = pd.read_excel(f"Daten/Schulferien/{filename}")
  # Dataframe mit entfernten Leerzeichen am Ende
  df_modified = remove_trailing_spaces(df.copy())
  # Ausgabe des modifizierten Dataframes
  df_modified.to_csv(f"Daten/Schulferien/{filename[:-5]}.csv", index=False)
```
