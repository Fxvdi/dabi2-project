```python
import os
import pandas as pd

# Verzeichnis definieren
directory = '../Data_Lake/Historisierung/Wetterdaten/Tagesbasis'

# Liste zum Speichern der DataFrames
dataframes = []

# Durch alle Dateien im Verzeichnis iterieren
for filename in os.listdir(directory):
    # Vollständigen Pfad zur Datei bekommen
    file_path = os.path.join(directory, filename)
    
    # Überprüfen, ob es sich um eine Datei handelt
    if os.path.isfile(file_path):
        # Datei in DataFrame einlesen, hier wird angenommen, dass es sich um CSV-Dateien handelt
        df = pd.read_csv(file_path)
        
        # DataFrame zur Liste hinzufügen
        dataframes.append(df)
```
